```python
19.Yeni bir dashboard yaradın (15,15 ölçüsündə). Ən önəmli gördüyünüz 4 ədəd vizualı bu dashboard-da göstərin
```

```python
df['Netflix Release Date']=pd.to_datetime(df['Netflix Release Date'])
df['il']=df['Netflix Release Date'].dt.year
df['ay']=df['Netflix Release Date'].dt.month
df['rüb']=df['Netflix Release Date'].dt.quarter
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
dd= df.groupby(['il','ay'])['Viewership Score'].idxmax()
monthly_top=df.loc[dd]
sns.barplot(x=monthly_top['Title'],y=monthly_top['Viewership Score'])
plt.title('Aylara görə ən məşhur filmlər')
plt.subplot(2,2,2)
rüb_mean=df.groupby('rüb')['Viewership Score'].mean().reset_index()
sns.barplot(x = rüb_mean['rüb'],y=rüb_mean['Viewership Score'],color='green')
plt.title('Rüblərə görə orta izləmə skoru')
plt.subplot(2,2,4)
top5=df.sort_values(by='Viewership Score',ascending=False).head(5)
sns.barplot(x=top5['Viewership Score'],y=top5['Title'],color='orange')
plt.title('Ən çox izlənən 5 film')
```
